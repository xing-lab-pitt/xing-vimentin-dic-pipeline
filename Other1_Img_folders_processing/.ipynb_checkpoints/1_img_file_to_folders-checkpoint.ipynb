```python
def folder_pattern(folder, start_posi, end_posi):
    """
    Return a list of patterns.
    Notice the actual index is (start - 1, end)
    """
    img_files = util.folder_file_num(folder)
    pattern_list = []
    i = 0
    while i<len(img_files):
        base_name = os.path.basename(img_files[i])
        pattern = base_name[int(start_posi-1): int(end_posi)]
        pattern_list.append(pattern)
        
        i = i+1
    pattern_list = np.unique(pattern_list)
    return pattern_list
```

```python
folder_pattern(img_folder, 14, 18)
```

```python
def to_subfolder(start_folder, end_folder, pattern, num_of_files = -1):
    """
    moving files in starting_folder to end_folder according to patterns
    """
    start_folder = util.folder_verify(start_folder)
    end_folder = util.folder_verify(end_folder)
    util.create_folder(end_folder)
    i = 0
    while i< len(pattern):
        
        p = pattern[i]    # current pattern
        d = util.folder_verify(end_folder) + p# current folder
        util.create_folder(d)
        util.file_copy(start_folder, p, num_of_files, d)
        i = i+1
    print("File move completed, please check your folder")
```

```python
pattern = folder_pattern(img_folder, 14, 18)
aim_folder = util.folder_verify(img_folder)[:-1] + "-sorted"
to_subfolder(img_folder, aim_folder, pattern)
```

```python
img_folder = "/net/capricorn/home/xing/huijing/Segmentation/data/5-28-21-fast-imaging/"
pattern = folder_pattern(img_folder, 7, 11)
aim_folder = util.folder_verify(img_folder)[:-1] + "-sorted"
to_subfolder(img_folder, aim_folder, pattern)
```
