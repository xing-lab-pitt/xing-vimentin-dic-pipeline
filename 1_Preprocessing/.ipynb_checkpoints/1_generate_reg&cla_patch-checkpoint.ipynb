```python
# The dimention of the croped pics
# This function works for mac and linux dir
# Use the picture size to decide proper patch size for crop. 
# Return the total number of ori images (set)
def image_sizes(folder):
    if "/" ==folder[-1]:
        pass
    else:
        folder = folder + "/"
    files = glob.glob(folder+"*.png")
    i = 0 
    for f in files:
        
        if "colony_" in f:
            img = imread(f, as_gray=True)
            print(img.shape, img.max(), img.min())
            
            i = i+1 # count total number of pictures
    print("Totally %s pics were croped."%i)
    
    return i

def small_img_remove(folder, thresh = 128):
    """
    loop over a folder. 
    if a image size is too small, then remove the file from the folder.
    """
    if "/" ==folder[-1]:
        pass
    else:
        folder = folder + "/"
        
    files = glob.glob(folder+"*.tif")
    remove_str = []
    
    i = 0 
    for f in files:
        
        if "crop_" in f:
            img = imread(f, as_gray=True)
            
            if (img.shape[0]<thresh or img.shape[1]<thresh):
                img_str = os.path.basename(f)[5:-4]
                remove_str.append(img_str)
                #print(img_str)
                continue
                
            print(img.shape, img.max(), img.min())
            
            i = i+1 # count total number of pictures
    print("Totally %s pics were croped."%i)
    
    files = glob.glob(folder + "*")
    for f in files:
        for s in remove_str:
            if s in f:
                os.remove(f)
                #print(f)
    return i
```

```python
#------weight map parameter----------
sigma=30
w_thres=np.exp(-4)
gamma=2

wv=1/np.exp(-2)


def cal_weights_map(Interior):
    
    Interior=np.asarray(Interior)
    if np.amax(Interior)>0:
        seg=label(Interior) # labeling connected blocks

        edt1=distance_transform_edt(seg==0) # Euclidean distance transformation. Then the pixel that far away from 0s will have highest values.
        edt2=np.zeros((int(np.amax(seg)),seg.shape[0],seg.shape[1]))
        for i in range(1,np.amax(seg)+1):
            other_seg=seg.copy()
            other_seg[other_seg==i]=0
            edt2[i-1,:,:]=distance_transform_edt(other_seg==0) # blocks euclidean distance

        edt=edt1+np.amax(edt2,axis=0) # combining 

        bnd_weights=np.exp(-(edt/sigma)**gamma)*(seg==0)
        bnd_weights[bnd_weights<w_thres]=0

        norm_bnd_weights=bnd_weights/(max(np.amax(bnd_weights),1/wv)+1e-7)
        bnd_weights=bnd_weights*wv#np.median(bnd_weights[bnd_weights>0])


        gt=distance_transform_edt(Interior)
        if np.count_nonzero(gt)!=0:
            nonzero_gt=gt[gt>0]
            median_edt=gt*1.0/np.median(nonzero_gt)
            norm_edt=gt*1.0/np.amax(nonzero_gt)
         
        median_weights_map=bnd_weights-median_edt
        norm_weights_map=norm_bnd_weights-norm_edt
    else:
        median_weights_map,norm_weights_map=np.zeros(Interior.shape),np.zeros(Interior.shape)
        
    median_wp=PImage.fromarray(median_weights_map)
    norm_wp=PImage.fromarray(norm_weights_map)
    #print(np.amax(norm_wp,axis = 0), norm_wp.size)
    return median_wp,norm_wp
```

```python
#generating weight map from the function above
#median weight map and the norm weight map 
Img_str=sorted(glob.glob(ori_path+ "crop*.tif"))
Interior_str=sorted(glob.glob(ori_path+ "interior*.png"))

for ti in range(len(Img_str)):
    #print(Img_str[ti])
    Interior=PImage.open(Interior_str[ti])
    median_wp,norm_wp = cal_weights_map(Interior)
    norm_wp_path = Img_str[ti].replace("crop", "norm_weights_map")
    median_wp_path = Img_str[ti].replace("crop", "median_weights_map")
    #print(norm_wp_path)
    norm_wp.save(norm_wp_path)
    median_wp.save(median_wp_path)
```

```python
#random crop image into patches

Img_str=sorted(glob.glob(ori_path+ "crop*.tif"))

Boundary_str=sorted(glob.glob(ori_path+ "boundary*.png"))
Interior_str=sorted(glob.glob(ori_path+ "interior*.png")) # binary, 0 or 1
BIB_str=sorted(glob.glob(ori_path+ "BIB*.png")) # background, interior, boundary, min 0, max 2.
#Norm_wp_str=sorted(glob.glob(ori_path+ "norm_weights_map*.tif"))
Bwdist_str=sorted(glob.glob(ori_path+ "bwdist*.png"))
Colony_str = sorted(glob.glob(ori_path+ "colony*.png"))

# set_: segmented interior, the brightness of each segment is different. 
# colony_: is the same with the interior.
```

```python
def image_trim(img_path, pixels = 3):
    """
    For a image path, trim several pixels on the boundary. 
    Pixels: a number 
    """
    img = PImage.open(img_path)
    img_w=img.size[0]
    img_h=img.size[1]
    img = img.crop((pixels, pixels, img_w - pixels, img_h - pixels))
    return img
    

train_patch_count=0
test_patch_count=0

for ti in np.arange(len(Img_str)):
    print(ti)
    Img      = image_trim(Img_str[ti])
    Boundary = image_trim(Boundary_str[ti])
    Interior = image_trim(Interior_str[ti])
    BIB      = image_trim(BIB_str[ti])
    #Norm_wp  = image_trim(Norm_wp_str[ti])
    Colony   = image_trim(Colony_str[ti])
    
    img_w=Img.size[0]
    img_h=Img.size[1]
    


    i=0
    j=0
    
    if ti in ti_train:
        #-------------------------------generate train patch
        train_patch=round((img_w*1.0/train_patch_w)*(img_h*1.0/train_patch_h)*crop_copy)

        while (i<train_patch):
            
            # if the croped picture is smaller than the patch dimention
            # disgard and go to the next picture. 
            if img_w<train_patch_w or img_h<train_patch_h:
                i = i+1
                continue
            
            xmin=round(random.uniform(0, 1)*(img_w-train_patch_w))
            ymin=round(random.uniform(0, 1)*(img_h-train_patch_h))
            rect=[xmin,ymin,xmin+train_patch_w,ymin+train_patch_h]
            #print(np.amax(np.asarray(Interior.crop(rect))))
#             if np.amax(np.asarray(Interior.crop(rect)))>0:
            train_patch_count+=1

            crop_Img=Img.crop(rect)
            crop_Img.save(train_img_path+'s'+str(train_patch_count)+'.tif');            

            crop_Interior=Interior.crop(rect)
            crop_Interior.save(train_interior_path+'s'+str(train_patch_count)+'.png');

#           crop_Bwdist=Bwdist.crop(rect)
            crop_Bwdist=distance_transform_edt(crop_Interior)
            crop_Bwdist=PImage.fromarray(crop_Bwdist)
            crop_Bwdist.save(train_reg_path+'s'+str(train_patch_count)+'.tif')

            crop_Boundary=Boundary.crop(rect);
            crop_Boundary.save(train_boundary_path+'s'+str(train_patch_count)+'.png');


            crop_BIB=BIB.crop(rect)
            crop_BIB.save(train_BIB_path+'s'+str(train_patch_count)+'.png')

#             crop_Median_wp,crop_Norm_wp=cal_weights_map(crop_Interior)
#             crop_Median_wp.save(train_mwp_path+'s'+str(train_patch_count)+'.tif')
            #crop_Norm_wp=Norm_wp.crop(rect)
            #crop_Norm_wp.save(train_nwp_path+'s'+str(train_patch_count)+'.tif')
        
            crop_Colony=Colony.crop(rect);
            crop_Colony_str=Colony_str[ti].replace(ori_path,'')
            crop_Colony.save(train_mask_path+crop_Colony_str[0:len(crop_Colony_str)-4]+'s'+str(train_patch_count)+'.png');

            i+=1
            
    else:
        #-----------------------------------generate test patch
        test_patch=round((img_w*1.0/test_patch_w)*(img_h*1.0/test_patch_h)*crop_copy)
        
        while (j<test_patch):
            
            # if the croped picture is smaller than the patch dimention
            # disgard and go to the next picture. 
            if img_w<test_patch_w or img_h<test_patch_h:
                j = j+1
                continue
                
            xmin=round(random.uniform(0, 1)*(img_w-test_patch_w))
            ymin=round(random.uniform(0, 1)*(img_h-test_patch_h))
            rect=[xmin,ymin,xmin+test_patch_w,ymin+test_patch_h]
            test_patch_count = test_patch_count+1
            
            crop_Img=Img.crop(rect)
            crop_Img_str=Img_str[ti].replace(ori_path,'')
            crop_Img.save(test_img_path+'s'+str(test_patch_count)+'.tif')   
            # print(test_img_path+'s'+str(test_patch_count)+'.tif')
            
            crop_Boundary=Boundary.crop(rect)
            crop_Boundary_str=Boundary_str[ti].replace(ori_path,'')
            crop_Boundary.save(test_boundary_path+'s'+str(test_patch_count)+'.png')
            
            crop_Interior=Interior.crop(rect)
            crop_Interior_str=Interior_str[ti].replace(ori_path,'')
            crop_Interior.save(test_interior_path+'s'+str(test_patch_count)+'.png')

         
            crop_Bwdist=distance_transform_edt(crop_Interior)
            crop_Bwdist_str=Bwdist_str[ti].replace(ori_path,'')            
            crop_Bwdist=PImage.fromarray(crop_Bwdist)
            crop_Bwdist.save(test_reg_path+'s'+str(test_patch_count)+'.tif')
            
            crop_BIB=BIB.crop(rect)
            crop_BIB_str=BIB_str[ti].replace(ori_path,'')
            crop_BIB.save(test_BIB_path+'s'+str(test_patch_count)+'.png')
            
            crop_Colony=Colony.crop(rect);
            crop_Colony_str=Colony_str[ti].replace(ori_path,'')
            crop_Colony.save(test_mask_path+crop_Colony_str[0:len(crop_Colony_str)-4]+'s'+str(test_patch_count)+'.png');
            
            j+=1
```

# Look at the pictures generated

```python
def folder_img_show(folder):
    
    """
    A convient function for print pictures under a folder.
    20 pics at a time.
    """
    fig = plt.figure(figsize=(20, 20))
    img_path_list = glob.glob(folder + "/*")
    
    if len(img_path_list)<20:
        print("Not enough pictures to print")
        return 1
    
    for i,img_path in enumerate(img_path_list[:20]):
        fig.add_subplot(4,5, i+1)
        plt.imshow(imread(img_path)/256.)

    plt.show()
```
